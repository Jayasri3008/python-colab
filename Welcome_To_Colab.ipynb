```
import matplotlib.pyplot as plt


xpoints = [5,6,7,8,9,10]
x1points = [6,7,8,9,10,11]
ypoints = [12,13.5,15.5,17,13,6]
y2points = [6,8,10.5,13,15,7]

plt.plot(xpoints,ypoints,label="line one",color="red")
plt.plot(x1points,y2points,label="line one",color="violet")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.show()
```
