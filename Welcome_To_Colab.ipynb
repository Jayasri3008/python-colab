```
#basic graph
import matplotlib.pyplot as plt


x_axis = [1,2,3,4,5]
y_axis = [1,2,3,4,5]

plt.plot(x_axis,y_axis)
plt.title("A very basic graph")

plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")

plt.show()
```

```
import matplotlib.pyplot as plt

font_dict = {
    'family':'serif',
    'color': 'red',
    'weight':'bold',
    'size':14,
    'style':'italic'

}

x_axis = [1,2,3,4,5]
y_axis = [1,2,3,4,5]

plt.plot(x_axis,y_axis)
plt.title("A very basic graph",fontdict=font_dict)

plt.xlabel("X AXIS",fontdict=font_dict)
plt.ylabel("Y AXIS",fontdict=font_dict)

plt.xticks([1,2,3,4,5])
plt.yticks([1,2,3,4,5])
plt.show()
```

```
from matplotlib import rcParams
import matplotlib.pyplot as plt

rcParams['font.size'] = 14
rcParams['figure.figsize'] = [8,6]
rcParams['axes.labelsize'] = 12
rcParams['axes.titlesize'] = 16
rcParams['lines.linewidth'] = 2

plt.plot([1,2,3],[4,5,6])
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```
import matplotlib.pyplot as plt

font_dict = {
    'family':'serif',
    'color': 'red',
    'weight':'bold',
    'size':14,
    'style':'italic'

}
x_axis = [1,2,3,4,5]
y_axis = [1,2,3,4,5]
y2=[0,1,4,9,16]

plt.plot(x_axis,y_axis, 'bo--', label='1*1')
plt.plot(y_axis,y2, 'r.--', label='x**2')

plt.plot(x_axis[2:],y2[2:],'r.-')
plt.title("A very basic graph",fontdict=font_dict)
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.legend()

plt.show()

```
